```python
import pandas as pd
import numpy as np
import os
```

```python
temp_path_dir = "/home/fabihat/phd_quaid/2603_dl_prj/data/temp"
new_path_dir = "/home/fabihat/phd_quaid/2603_dl_prj/data/atac_chr19_subsampled"
```

```python
unique_id = sorted(set([i.split(".")[0] for i in os.listdir(temp_path_dir)]))
```

```python
unique_id
```

```python
for i in range(len(unique_id)):
    df_full = pd.read_csv(f"{temp_path_dir}/{unique_id[i]}.chr19_bins_full.tsv.gz", sep="\t")
    df_sparse = pd.read_csv(f"{temp_path_dir}/{unique_id[i]}.chr19_bins_sparse.tsv.gz", sep="\t")
    df = df_full.merge(df_sparse)
    df.to_csv(f"{new_path_dir}/{unique_id[i]}.chr19_bins.tsv.gz", sep="\t", index=False)
```
